```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = open("results_connection strength.json", "r")
res = f.read()
respd = pd.read_json(res, orient='table')
```

```python
def add_apendix_to_colnames(df, append):
    colnames = df['Percentage'].unstack().columns
    new_colnames =[]
    for colname in colnames:
        new_colnames.append(colname + ' ' + append)
    df2 = df['Percentage'].unstack()
    df2.columns = new_colnames
    df3 = df2.unstack()
    return df3
```

```python
respd = add_apendix_to_colnames(respd, 'connection')
respd
```

```python
f = open("results_cooperate.json", "r")
res = f.read()
respd2 = add_apendix_to_colnames(pd.read_json(res, orient='table'), 'cooperate')
respd2
```

```python
respd = respd.append(respd2)
respd
```

```python
f = open("results_fight.json", "r")
res = f.read()
respd2 = add_apendix_to_colnames(pd.read_json(res, orient='table'), 'fight')
respd2
```

```python
respd = respd.append(respd2)
respd
```

```python
resun = respd.unstack(0)
resun
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

appendix = 'all'
path = f'figures/correlation_{appendix}'

import pathlib
pathlib.Path(path).mkdir(parents=True, exist_ok=True)

methods = ['pearson', 'kendall', 'spearman']
for method in methods:
    g = sn.heatmap(resun.corr(method=method),annot= True)
    title = f"{method}".capitalize() + \
            f" correlation 2015-2020"
    g.set_title(title)
    g.figure.set_size_inches(15, 6)
    plt.savefig(f'{path}/{title}.png', bbox_inches='tight')
    plt.show()

```
